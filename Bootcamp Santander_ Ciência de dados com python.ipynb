```python
# API para manipulação de cadastro
scd2023_url_api = 'https://sdw-2023-prd.up.railway.app'
```

```python
# Etapa de extração dos IDs do arquivo SCD2023.csv e importação do pandas para manipular o arquivo
# Link do arquivo CSV: https://drive.google.com/file/d/1O9Vm-MjSik9AKJl5J6JEqpc5bUgId-k4/view?usp=drive_link
import pandas as pd

df = pd.read_csv('SCD2023.csv')
user_ids = df['UserID'].tolist()
print(user_ids)
```

```python
# Puxando ID de cada usuario através da API que foi criado o cadastro
# Importando bibliotecas para fazer chamadas e para manipulação dos dados
import requests
import json

# Função para puxar o json da API com os dados do usuario, caso contrário irá retornar None
def get_user(id):
  response = requests.get(f'{scd2023_url_api}/users/{id}')
  return response.json() if response.status_code == 200 else None

# Atribuindo o usuario em users caso o retorno não for None
users = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(users, indent=4))
```

```python
# Importando a biblioteca que instalamos anteriormente
import openai

openai.api_key = openai_api_key

# Função para gerar as mensagens para os usuarios
def generate_ai_news(user):
  completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Você é um especialista em marketing bancário no Santander."},
    {"role": "user", "content": f"Crie uma mensagem para {user['name']} sobre a importância dos investimentos (Máximo de 100 caracteres)"}
  ]
)
  return completion.choices[0].message.content.strip('\"')

# Passando as mensagens geradas para cada usuario
for user in users:
  news = generate_ai_news(user)
  print(news)
  user['news'].append({
      "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
      "description": news
  })
```

```python
# Pegando os dados gerados e passando pra API que foi criado os usuarios
def update_user(user):
  response = requests.put(f"{scd2023_url_api}/users/{user['id']}", json=user)
  return True if response.status_code == 200 else False

for user in users:
  sucess = update_user(user)
  print(f"User {user['name']} updated? {sucess}!")
```
